```python
#PyBer - Homework 5 - Matplotlib

# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Get the data files! (2 .csv files)
CityDataFile = pd.read_csv("Raw_data/city_data.csv") 
CityDataFile.head(7)
```

```python
# and the other file...
RideDataFile = pd.read_csv("Raw_data/ride_data.csv") 
RideDataFile.head(7)
```

```python
# Your objective is to build a Bubble Plot that showcases the relationship between four key variables:
# - Average Fare ($) Per City
# - Total Number of Rides Per City
# - Total Number of Drivers Per City
# - City Type (Urban, Suburban, Rural)

# data for the scatter plot:
# city, city type, driver count, rides.sum, revenue.sum

# let's build the result set!
CityCounts = CityDataFile['city'].value_counts()

# Using GroupBy in order to separate the data into fields according to 'city' values
GroupedByCity = CityDataFile.groupby(['city'])
# In order to be visualized, a data function must be used...
TotalDriversPerCity = GroupedByCity['driver_count'].sum()

# create an Output table
TableForScatter = pd.DataFrame({'CityID': CityCounts, 'Drivers': TotalDriversPerCity})

# convert the index into a column
TableForScatter.reset_index(inplace=True)
TableForScatter = TableForScatter.rename(columns = {'index' : 'city'})
TableForScatter.head(11)
```

```python
# merge the tables
MergedTableForScatter = pd.merge(TableForScatter, CityDataFile, on='city', how='inner')

MergedTableForScatter.head(7)
```

```python
# Manipulate the rides data / table

# Using GroupBy in order to summarize the data by 'city' values
GroupedByRideData = RideDataFile.groupby(['city'])

# Ride revenue, count and average revenue
RevByCity = GroupedByRideData['fare'].sum()
RideCount = RideDataFile['city'].value_counts()
AveRideRev = round((RevByCity / RideCount), 2)
#print(AveRideRev)

# Construct the output table
# In order to be visualized, a data function must be used...
RideOutputTable = pd.DataFrame({'Total Rev': RevByCity,
                                'Ride Count': RideCount,
                                'Average Ride Rev': AveRideRev})
# convert the index into a column
RideOutputTable.reset_index(inplace=True)
RideOutputTable = RideOutputTable.rename(columns = {'index' : 'city'})
RideOutputTable.head(7)

```

```python
#Time to merge the tables!!!
MergedCityAndRevTable = pd.merge(MergedTableForScatter, RideOutputTable, on='city', how='inner')
MergedCityAndRevTable = MergedCityAndRevTable.drop('driver_count', axis=1)
MergedCityAndRevTable = MergedCityAndRevTable.drop('CityID', axis=1)
MergedCityAndRevTable.head(7)
```

```python
#Create a lookup for typ with color and size
# Labels for the sections of the Scatter chart
ScatterTypes = ["Rural", "Urban", "Suburban"]

# The colors of each section of the Scatter chart
ScatterColors = ['gold', 'lightcoral', 'Lightskyblue']

# bubble sizes
BubbleSizes = [25, 150, 75]

LookUpColumnNames = ['type', 'Bubble Color', 'Bubble Size']

LookupTyp_df = pd.DataFrame(columns=LookUpColumnNames) # Note that there are no row data inserted.
LookupTyp_df['type'] = ScatterTypes
LookupTyp_df['Bubble Color'] = ScatterColors
LookupTyp_df['Bubble Size'] = BubbleSizes

LookupTyp_df.head()
```

```python
# make a copy of the table
ScatterPlotEnh = MergedCityAndRevTable
ScatterPlotEnh.head(7)

# Merge with the lookup table
MergedScatterPlotEnh = pd.merge(ScatterPlotEnh, LookupTyp_df, on='type', how='inner')
MergedScatterPlotEnh.head(11)
```

```python
# Time for a chart!

UrbanScatterPlot= MergedScatterPlotEnh.loc[MergedScatterPlotEnh['type'] == 'Urban',:]
SuburbanScatterPlot= MergedScatterPlotEnh.loc[MergedScatterPlotEnh['type'] == 'Suburban',:]
RuralScatterPlot= MergedScatterPlotEnh.loc[MergedScatterPlotEnh['type'] == 'Rural',:]
#RuralScatterPlot.head(11)

#Urban
x_axis = UrbanScatterPlot['Ride Count']
y_axis = UrbanScatterPlot['Average Ride Rev']
UrbanScatter =  plt.scatter(x_axis, y_axis, marker="o", facecolors= UrbanScatterPlot['Bubble Color'], edgecolors="black",
            s=UrbanScatterPlot['Bubble Size'], alpha=0.40)

#Suburban
x_axis = SuburbanScatterPlot['Ride Count']
y_axis = SuburbanScatterPlot['Average Ride Rev']
SuburbanScatter =  plt.scatter(x_axis, y_axis, marker="o", facecolors= SuburbanScatterPlot['Bubble Color'], edgecolors="black",
            s=SuburbanScatterPlot['Bubble Size'], alpha=0.50)

#Rural
x_axis = RuralScatterPlot['Ride Count']
y_axis = RuralScatterPlot['Average Ride Rev']
RuralScatter =  plt.scatter(x_axis, y_axis, marker="o", facecolors= RuralScatterPlot['Bubble Color'], edgecolors="black",
            s=RuralScatterPlot['Bubble Size'], alpha=0.6)

plt.title("Total Fares by City Type")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare")

# Set our legend to where the chart thinks is best
plt.legend([UrbanScatter, SuburbanScatter, RuralScatter], ['Urban', 'Suburban', 'Rural'])
```

```python
# In addition, you will be expected to produce the following three pie charts:
# - % of Total Fares by City Type
# - % of Total Rides by City Type
# - % of Total Drivers by City Type

# Group by City Type
# Using GroupBy in order to summarize the data by 'city' values
GroupedByCityType = MergedCityAndRevTable.groupby(['type'])

# Fares, Rides, and Drivers by City type
TotalFaresByType = GroupedByCityType['Total Rev'].sum()
RidesByCityType = GroupedByCityType['Ride Count'].sum()
DriversByCityType = GroupedByCityType['Drivers'].sum()

# Construct the output table
# In order to be visualized, a data function must be used...
CityTypeOutputTable = pd.DataFrame({'Total Rides': RidesByCityType,
                                    'Total Fares': TotalFaresByType,
                                    'Total Drivers': DriversByCityType})

# convert the index into a column
CityTypeOutputTable.reset_index(inplace=True)
CityTypeOutputTable = CityTypeOutputTable.rename(columns = {'index' : 'city type'})
CityTypeOutputTable.head(7)
```

```python
# Pie chart time!  ("mmmmmmmmmmm pie!" - Homer Simpson)

# % of Total Fares by City Type

# Labels for the sections of our pie chart
PyberLabels = ["Rural", "Suburban", "Urban"]

# The colors of each section of the pie chart
PyberColors = ['gold', 'lightskyblue', 'Lightcoral']

# Tells matplotlib to seperate the "Suburban" section from the others
PyberExplode = (0, 0, 0.1)

fares = CityTypeOutputTable['Total Fares']
citytype = CityTypeOutputTable['type']
FarePie = fares.plot(kind = 'pie',
                     y = citytype, title = '% of Total Fares by City Type',
                     autopct="%1.1f%%",
                     explode = PyberExplode,
                     labels = PyberLabels,
                     colors = PyberColors,
                     shadow=True,
                     startangle=150)

plt.show()
```

```python
# % of Total Rides by City Type

rides = CityTypeOutputTable['Total Rides']
citytype = CityTypeOutputTable['type']
RidePie = rides.plot(kind = 'pie',
                     y = citytype, title = '% of Total Rides by City Type',
                     autopct="%1.1f%%",
                     explode = PyberExplode,
                     labels = PyberLabels,
                     colors = PyberColors,
                     shadow=True,
                     startangle=140)

plt.show()
```

```python
# % of Total Drivers by City Type

drivers = CityTypeOutputTable['Total Drivers']
citytype = CityTypeOutputTable['type']
DriverPie = drivers.plot(kind = 'pie',
                     y = citytype, title = '% of Total Drivers by City Type',
                     autopct="%1.1f%%",
                     explode = PyberExplode,
                     labels = PyberLabels,
                     colors = PyberColors,
                     shadow=True,
                     startangle=160)

plt.show()
```
